```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
```

```python
df = pd.read_excel("amazon.csv.xlsx")
df.head()
```

```python
print("Dataset Shape:", df.shape)

df.info()
df.describe()
df.columns
```

```python
df.drop_duplicates(inplace=True)
print("Shape after removing duplicates:", df.shape)
```

```python
df.isnull().sum()
```

```python
#removing certain columns which are not useful for analytics dashboards
```

```python
columns_to_drop = ['img_link','product_link','user_id','user_name','review_id','review_content']

df.drop(columns=columns_to_drop, inplace=True, errors='ignore')
df.columns
```

```python
#Cleaning the discounted and actual price - columns to perform calculations
```

```python
df['discounted_price'] = df['discounted_price'].replace('[^0-9.]', '', regex=True)
df['actual_price'] = df['actual_price'].replace('[^0-9.]', '', regex=True)

df['discounted_price'] = df['discounted_price'].astype(float)
df['actual_price'] = df['actual_price'].astype(float)
```

```python
#Cleaning the rating count
```

```python
df['rating_count'] = df['rating_count'].replace(',', '', regex=True)
df['rating_count'] = df['rating_count'].astype(float)
```

```python
#Cleaning rating columns
```

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['rating_count'] = df['rating_count'].replace(',', '', regex=True)
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')
```

```python
df[['rating','rating_count']].head()
```

```python
df['main_category'] = df['category'].str.split('|').str[0]
```

```python
#Estimating revenue based on rating count as real sales quantity is not available
```

```python
df['estimated_revenue'] = df['discounted_price'] * df['rating_count']
```

```python
#EXPLORATORY DATA ANALYSIS ON :
```

```python
#1. TOP CATEGORIES BY REVENUE
```

```python
category_revenue = df.groupby('main_category')['estimated_revenue'].sum().sort_values(ascending=False)
print(category_revenue.head(10))
```

```python
#2. TOP PRODUCTS BY REVENUE
```

```python
top_products = df.sort_values(by='estimated_revenue', ascending=False).head(10)
print(top_products[['product_name','main_category','estimated_revenue']])
```

```python
#3. AVERAGE RATING BY CATEGORY
```

```python
category_rating = df.groupby('main_category')['rating'].mean().sort_values(ascending=False)
print(category_rating.head(10))
```
